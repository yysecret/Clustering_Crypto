```python
# Alternatively, use the provided csv file:
file_path = Path("crypto_data.csv")

# Create a DataFrame
response_df = pd.read_csv(file_path, index_col= 0)
response_df
```

### Data Preprocessing

```python
# Keep only necessary columns:
# 'CoinName','Algorithm','IsTrading','ProofType','TotalCoinsMined','TotalCoinSupply'
crypto_df = response_df[['CoinName','Algorithm','IsTrading','ProofType','TotalCoinsMined','TotalCoinSupply']]
crypto_df
```

```python
# Keep only cryptocurrencies that are trading
crypto_df = crypto_df[crypto_df['IsTrading']== True]
crypto_df

```

```python
# Keep only cryptocurrencies with a working algorithm
crypto_df = crypto_df[crypto_df['Algorithm'] != 'N/A']
crypto_df
```

```python
# Remove the "IsTrading" column
crypto_df = crypto_df.drop(columns='IsTrading')
crypto_df
```

```python
# Remove rows with at least 1 null value
crypto_df = crypto_df.dropna()
crypto_df.isnull().sum()
```

```python
# Remove rows with cryptocurrencies having no coins mined
crypto_df = crypto_df[crypto_df['TotalCoinsMined'] > 0]
crypto_df
```

```python
# Drop rows where there are 'N/A' text values
crypto_df = crypto_df[(crypto_df['CoinName'] != 'N/A') | (crypto_df['ProofType'] != 'N/A') ]
crypto_df.head()
```

```python
# Store the 'CoinName'column in its own DataFrame prior to dropping it from crypto_df
coins_name = pd.DataFrame(data = crypto_df, columns=['CoinName'])
coins_name.head()
```

```python
# Drop the 'CoinName' column since it's not going to be used on the clustering algorithm
crypto_df.drop(columns='CoinName', inplace=True)
crypto_df.head()
```

```python
# Create dummy variables for text features
X = pd.get_dummies(crypto_df, columns=['Algorithm','ProofType'])
```

```python
# Standardize data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled
```

### Reducing Dimensions Using PCA

```python
# Use PCA to reduce dimensions to 3 principal components
pca = PCA(n_components=3)
```

```python
# Create a DataFrame with the principal components data
X_pca = pca.fit_transform(X_scaled)
pca_df = pd.DataFrame(X_pca, index=crypto_df.index)
pca_df.columns = ['PC1','PC2','PC3']
pca_df
```

### Clustering Crytocurrencies Using K-Means

#### Find the Best Value for `k` Using the Elbow Curve

```python
inertia = []
k = list(range(1, 11))

# Calculate the inertia for the range of k values
for i in k:
    k_model = KMeans(n_clusters=i, random_state=1)
    k_model.fit(pca_df)
    inertia.append(k_model.inertia_)

# Define a DataFrame to hold the values for k and the corresponding inertia
elbow_data = {"k": k, "inertia": inertia}
df_elbow = pd.DataFrame(elbow_data)

# Review the DataFrame
df_elbow
```

```python
# Create the Elbow Curve using altiar
alt.Chart(df_elbow).mark_line(point=True).encode(
    x=alt.X('k'),
    y=alt.Y('inertia')
)
```

Running K-Means with `k=<your best value for k here>`

```python
# Initialize the K-Means model
model = KMeans(n_clusters=4, random_state=1)
# Fit the model
model.fit(pca_df)
# Predict clusters
predictions = model.predict(pca_df)
# Create a new DataFrame including predicted clusters and cryptocurrencies features
predicted_cluster_df = crypto_df.copy()

# Add a class column with the labels
predicted_cluster_df['class'] = predictions
clustered_df = pd.concat([coins_name,predicted_cluster_df,pca_df],axis='columns')
clustered_df.head(10)
```

### Visualizing Results

#### Scatter Plot with Tradable Cryptocurrencies

```python
# Scale data to create the scatter plot
scaler = MinMaxScaler()
scaled = scaler.fit_transform(clustered_df[['TotalCoinsMined', 'TotalCoinSupply']])
scaled_clustered_df = clustered_df.copy()
scaled_clustered_df[['TotalCoinsMined', 'TotalCoinSupply']] = scaled
scaled_clustered_df
```

```python
# Plot the scatter with x="TotalCoinsMined" and y="TotalCoinSupply"
points = alt.Chart(scaled_clustered_df).mark_circle().encode(
    x='PC1',
    y='PC2',
    color=alt.Color('class', scale=alt.Scale(domain=[0,1,2,3], range=['red','yellow','blue','green'])),
    tooltip=['CoinName','TotalCoinsMined', 'TotalCoinSupply']).interactive()

points
```
